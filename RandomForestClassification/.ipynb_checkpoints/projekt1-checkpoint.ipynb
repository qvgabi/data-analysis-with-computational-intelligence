```python
import matplotlib.pyplot as plt
import random
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("adult.csv")
```

```python
df.head()
```

```python
# data cleaning
df['income'] = df['income'].map({'<=50K': 0, '>50K': 1})
df = df.replace('?',np.nan)
df = df.drop(['native-country'], axis=1)
```

```python
# checking for columns with nan values
df.isna().any()
```

```python
# replacing missing values with most frequent values
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df['workclass'] = imp.fit_transform(df['workclass'].values.reshape(-1,1))
df['occupation'] = imp.fit_transform(df['occupation'].values.reshape(-1,1))
# df['native-country'] = imp.fit_transform(df['native-country'].values.reshape(-1,1))
df.head()
```

```python
income = df["income"]
income.value_counts()
```

```python
df.info()
```

```python
df.shape
```

```python
# more detailed info about data
for i in df.columns.tolist():
    print(i,':')
    print(df[i].value_counts())
    print('')

# generalizing data? for example education is a bit messy, 
# could replace with preschool, high school, higher ???
```

```python
# preparing data (encoding)
from sklearn.preprocessing import OrdinalEncoder

data = df.drop('income',axis=1)
obj_df = data.select_dtypes(include=['object']).copy()

ord_enc = OrdinalEncoder()

for i in obj_df:
    data[i] = ord_enc.fit_transform(obj_df[[i]])
    
data.head()
```

```python
# classification: random forest
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier


X_train, X_test, y_train, y_test = train_test_split(data,income, train_size = 0.7, test_size = 0.3, random_state = 0)

# to do:  zabawa z hiperparametrami
model = RandomForestClassifier(n_estimators = 50, criterion = "entropy", max_depth = 20, 
                               min_samples_split = 2, min_samples_leaf = 3, min_weight_fraction_leaf = 0.0,
                               max_features = "auto")
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("Accuracy on test set: ", model.score(X_test,y_test)*100,"%")
print("Accuracy on traing set : ", model.score(X_train,y_train)*100,"%")
```

```python
# confusion matrix

from sklearn.metrics import confusion_matrix,plot_confusion_matrix
conf=confusion_matrix(y_true=y_test,y_pred=y_pred)
print(conf)

plot_confusion_matrix(model,X_test,y_test)
plt.show()
```

```python
# cross validation ?? szczerze mówiąc to nie wiem co się tutaj dzieje, muszę poczytać
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, data , income, cv=10)
print(scores)
```

```python
# tuning hyperparameters, tez nie wiem o chuj chodzi, niby jest to grid_search ale chuj wie

from sklearn.model_selection import train_test_split,cross_val_score,KFold,GridSearchCV

model = RandomForestClassifier()
kf = KFold(n_splits=3)
max_features = np.array([1,2,3,4,5])
n_estimators = np.array([25,50,75,100])
min_samples_leaf=np.array([25,50,75,100])
param_grid=dict(n_estimators=n_estimators,max_features=max_features,min_samples_leaf=min_samples_leaf)
grid=GridSearchCV(estimator=model,param_grid=param_grid,cv=kf)
gres=grid.fit(X_train,y_train)
print("Best",gres.best_score_)
print("params",gres.best_params_)
```
